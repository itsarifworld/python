### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line



```python
class Point:

  # constructor
  def __init__(self, x, y):
    self.x_cor = x
    self.y_cor = y

  def __str__(self):
    return '<{},{}>'.format(self.x_cor, self.y_cor)


  def euclidean_distance(self, other):
    print("Distance Between <{},{}> and <{},{}> is = ".format(self.x_cor, self.y_cor, other.x_cor, other.y_cor))
    return ((self.x_cor - other.x_cor)**2 + (self.y_cor - other.y_cor)**2)**0.5

  def origin_distance(self):
    print("Distance Between <{},{}> and <0,0> is = ".format(self.x_cor, self.y_cor))
    #return ((self.x_cor)**2 + (self.y_cor)**2)**0.5
    return self.euclidean_distance(Point(0,0))
```

```python
p1 = Point(0,0)
p2 = Point(3,4)
print(p1)
print(p2)
```

```python
class Line:

  def __init__(self, A, B, C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return '{}x + {}y + {} = 0'.format(self.A, self.B, self.C)

  def point_on_line(line, point):
    if line.A*point.x_cor + line.B*point.y_cor + line.C == 0:
      return "Point <{},{}> lies on the line".format(point.x_cor,point.y_cor)
    else:
      return "Point <{},{}> does not lie on the line".format(point.x_cor,point.y_cor)

  def shortest_distance(line,point):
    return abs(line.A*point.x_cor + line.B*point.y_cor + line.C)/(line.A**2 + line.B**2)**0.5
```
